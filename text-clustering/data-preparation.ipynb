## Making df_chrea_obrist-ground_truth

```python
df_chrea_obrist = obrist[['Codes_X_ClusterOdorQuality']]
df_chrea_obrist.columns = ['cluster']
```

```python
df_cluster_community = pd.DataFrame(columns=['text', 'clusterID', 'communityID'], index=[x for x in range(len(df_chrea_obrist))])
df_cluster_community
```

```python
cluster_dict = {}
cluster_comm_dict = {}

def extract_cluster_community(clusterText):
    global cluster_dict, cluster_comm_dict
    splitted = clusterText.split("--")
    cluster, comm = splitted[0].strip(), splitted[1].strip()
    cluster_splitted = [s for s in cluster.split()]
    
    current_cluster = 0
    
    ## Update cluster_dict ##
    try:
        (cluster_splitted[1]).isdigit()
        cluster_id = cluster_splitted[1]
    except:
        raise Exception("cluster number is not a digit")
    
    cluster_meaning = cluster_splitted[2]
    
    if cluster_id not in cluster_dict:
        cluster_dict.update({cluster_id: cluster_meaning})
    else:
        # cluster_id exist in cluster_dict
        if cluster_meaning != cluster_dict[cluster_id]:
            # We do not expect this part to be executed
            raise Exception("same cluster number but different description?")
    
    current_cluster = cluster_id
    
    ## Update cluster_comm_dict ##
    if current_cluster not in cluster_comm_dict:
        cluster_comm_dict.update({current_cluster: {1: comm}})
        # comm_id looks like 1-1, 1-2 ...
        comm_id = f'{current_cluster}-1'
    else:
        comm_dict = cluster_comm_dict.get(current_cluster)
        if comm in comm_dict.values():
            # comm_id looks like 1-1, 1-2 ...
            comm_id = f'{current_cluster}-{list(comm_dict.keys())[list(comm_dict.values()).index(comm)]}'
        else: # comm not in comm_dict values
            new_comm_dict_key = len(comm_dict)+1
            cluster_comm_dict[current_cluster].update({new_comm_dict_key : comm})
            comm_id = f'{current_cluster}-{new_comm_dict_key}'
    
    
    return cluster_id, comm_id
```

```python
for index, row in df_chrea_obrist.iterrows():
    cluster_id, comm_id = extract_cluster_community((row.cluster).strip())
    df_cluster_community.iloc[index]['text'] = row.text
    df_cluster_community.iloc[index]['clusterID'] = cluster_id
    df_cluster_community.iloc[index]['communityID'] = comm_id
```
