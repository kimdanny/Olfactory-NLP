```
import nltk
import pandas as pd

from textblob import TextBlob

import spacy
!pip install spacytextblob
from spacytextblob.spacytextblob import SpacyTextBlob
nlp = spacy.load('en_core_web_sm')
spacy_text_blob = SpacyTextBlob()
nlp.add_pipe(spacy_text_blob)
!pip install text2emotion
import text2emotion as te

# HELPER FN FOR SUBJECTIVITY AND POLARITY
def detect_polarity(text):
    return TextBlob(text).sentiment


# GET A LIST OF ALL THE STORIES
stories = pd.read_csv('stories1.csv')

processedData = []


for element in stories['col']:
  processedData.append(element)

print(processedData)

sentences = processedData
vectors_df = pd.DataFrame(columns=['compound', 'neg', 'neu' , 'pos', 'polarity', 'subjectivity', 'Happy', 'Angry', 'Surprise', 'Sad', 'Fear'])


# GET A LIST OF ALL THE SENTIMENTS:
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
sid = SentimentIntensityAnalyzer()




index = 0
for sentence in sentences:
  ss = sid.polarity_scores(sentence)
  for k in sorted(ss):
    # print('{0}: {1}, '.format(k, ss[k]), end='')
    vectors_df.loc[index, k] = ss[k]
  # print()

  # result = detect_polarity(sentence)
  # # print(result)
  # vectors_df.loc[index, 'polarity'] = result.polarity
  # vectors_df.loc[index, 'subjectivity'] = result.subjectivity
  # print(result.polarity)
  # print(result.subjectivity)

  doc = nlp(sentence)
  vectors_df.loc[index, 'polarity'] = doc._.sentiment.polarity
  vectors_df.loc[index, 'subjectivity'] = doc._.sentiment.subjectivity
  
  # print(doc._.sentiment.polarity)
  # print(doc._.sentiment.subjectivity)
  # print(doc._.sentiment.assessments)

  emotions = te.get_emotion(sentence)

  for key in emotions:
    vectors_df.loc[index, key] = emotions[key]


  index = index + 1

# print("currentDT:")
print(vectors_df)


vectors_df.to_csv('sentiment_unproc.csv')
```
