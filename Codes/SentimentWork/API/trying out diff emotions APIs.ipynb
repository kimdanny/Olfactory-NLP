```
import pandas as pd


df = pd.DataFrame(columns = ['Story', 'joy', 'sadness', 'surprise', 'fear', 'anger', 'disgust']) 
df
```

```
storiesProcesses = pd.read_csv('stories1.csv')

processedData = []

for element in storiesProcesses['col']:
  processedData.append(element)

print(processedData)
```

```
import requests
import json
import requests

# 2dda301f0dmshbc22bde8cac0f49p1f80b6jsnf301d21422a6
# pls dont add the API key - imma be broke - 

def makeRequest(input, index):
  url = "https://twinword-emotion-analysis-v1.p.rapidapi.com/analyze/"

  querystring = {"text": input}

  headers = {
      'x-rapidapi-key': "",
      'x-rapidapi-host': "twinword-emotion-analysis-v1.p.rapidapi.com"
      }

  response = requests.request("GET", url, headers=headers, params=querystring)

  result = response.text

  df.loc[index, 'Story'] = processedData[index]
  json_data = json.loads(result)
  for key in json_data["emotion_scores"]:
    df.loc[index, key] = json_data["emotion_scores"][key]


for i in range(len(processedData)):
  print(i)
  makeRequest(processedData[i], i)

df.head()
```
